### Function to extract images from directory and saving in dictionary.
We have four different eras :   
'1996-2000'  
'2001-2004'  
'2005-2008'  
'2009-2013'   
We will try to classify our images and classify them into one of the four eras.

```python
def getFiles(path):

    imlist = {}
    imlist1 = {}
    count = 0
    for each in os.listdir(path):

        print (" #### Reading image category ", each, " ##### ")
        #imlist[each] = []
        path1 = path+'\\'+each
        
        for imagefile in os.listdir(path1):

            print ("Reading file ", imagefile)
            if imagefile[:4] not in imlist.keys():
                imlist[imagefile[:4]] = []

            im = cv2.imread(path1+'\\'+imagefile)
            #cv2.imshow('image',im)
            #imlist[each].append(im)
            im1 = cv2.resize(im,(32, 32), interpolation = cv2.INTER_CUBIC)
            #im2 = cv2.resize(im,(512, 512), interpolation = cv2.INTER_CUBIC)
            imlist[imagefile[:4]].append(im1)
            count +=1
            
    return [imlist, count]
```

Function for one hot encoding

```python
def one_hot_encode(vec):
    final_arr = []
    for each in vec:
        if each=="1996-2000":
            arr = [1,0,0,0]
        elif each == '2001-2004':
            arr = [0,1,0,0]
        elif each == '2005-2008':
            arr = [0,0,1,0]
        else:
            arr= [0,0,0,1]

        final_arr.append(arr)
    final_arr = np.array(final_arr)
    return final_arr
```

```python
y= one_hot_encode(y)
```

```python
x = tf.placeholder(tf.float32,shape = [None,32,32,3])
y_true = tf.placeholder(tf.float32,shape = [None,4])
```

```python
hold_prob = tf.placeholder(tf.float32)
```

```python
def init_weights(shape):
    init_random_dist = tf.random_normal(shape,stddev=0.1)
    return tf.Variable(initial_value=init_random_dist)
```

```python
def init_bias(shape):
    init_random_bias = tf.constant(0.1,shape=shape)
    return tf.Variable(initial_value=init_random_bias)
```

```python
def conv2d(x,W):
    return tf.nn.conv2d(x,W,strides=[1,1,1,1],padding = 'SAME')
```

```python
def max_pool_2by2(x):
    return tf.nn.max_pool(x,[1,2,2,1],strides = [1,2,2,1],padding = 'SAME')
```

```python
def convolutional_layer(x,shape):
    W = init_weights(shape)
    b = init_bias([shape[3]])
    return tf.nn.relu(conv2d(x,W) + b)
```

```python
def normal_full_layer(x,size):
    input_size = int(x.get_shape()[1])
    W = init_weights([input_size,size])
    b = init_bias([size])
    return tf.matmul(x,W)+b
```

```python
convo_1= convolutional_layer(x,shape = [4,4,3,32])
convo_1_pooling = max_pool_2by2(convo_1)
```

```python
convo_2 = convolutional_layer(convo_1_pooling,shape = [4,4,32,64])
convo_2_pooling = max_pool_2by2(convo_2)
```

```python
convo2_flattened = tf.reshape(convo_2_pooling,[-1,4096])
```

```python
full_layer = normal_full_layer(convo2_flattened,size = 1024)
last_layer = tf.nn.relu(full_layer)
```

```python
dropout_layer = tf.nn.dropout(last_layer, hold_prob)
```

```python
y_pred = normal_full_layer(dropout_layer,size = 4)
```

```python
cross_entropy = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y_true,logits=y_pred))
```

```python
optimizer = tf.train.AdamOptimizer(0.001)
train = optimizer.minimize(cross_entropy)
```
