### Function to extract images from directory and saving in dictionary.
We need to classify our image into one of the 26 different genres. 

```python
def getFiles(path):
    """
    - returns  a dictionary of all files 
    having key => value as  objectname => image path
    - returns total number of files.
    """
    
    imlist = {}
    count = 0
    for each in os.listdir(path):

        print (" #### Reading image category ", each, " ##### ")
        #imlist[each] = []
        path1 = path+'\\'+each
        
        for imagefile in os.listdir(path1):

            print ("Reading file ", imagefile)
            if each not in imlist.keys():
                imlist[each] = []

            im = cv2.imread(path1+'\\'+imagefile)
            #cv2.imshow('image',im)
            #imlist[each].append(im)
            im1 = cv2.resize(im,(227, 227), interpolation = cv2.INTER_CUBIC)
            imlist[each].append(im1)
            count +=1

    return [imlist, count]
```

```python
path = r"C:\Users\Surbhi\Desktop\Infoviz\Project\Genre"
img_dict,count = getFiles(path)
```

```python
X = []
y = []
for each_key in img_dict.keys():
    for every in img_dict[each_key]:
        X.append(every)
        y.append(each_key)
X = np.array(X)
y = np.array(y)
```

```python
X.shape
```

```python
img_dict_keys = list(img_dict.keys())
img_dict_keys_dictionary = {}

for i in range(len(img_dict_keys)):
    img_dict_keys_dictionary[img_dict_keys[i]] = np.zeros(len(img_dict_keys))
    img_dict_keys_dictionary[img_dict_keys[i]][i] = 1
```

```python
def one_hot_encode(vec):
    final_arr = []
    for each in vec:
        final_arr.append(list(img_dict_keys_dictionary[each]))
    final_arr = np.array(final_arr)
    return final_arr

y= one_hot_encode(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5,stratify =y )
```

```python
class Helper():
    def __init__(self):
        self.i = 0
        self.train_batch = X_train
        self.test_batch = X_test
        
        self.training_images = None
        self.training_labels = None
        
        self.test_images = None
        self.test_labels = None
        
    def set_images(self):
        self.training_images = X_train
        train_len = len(self.training_images)
        self.training_images = self.training_images.reshape(train_len,3,32,32).transpose(0,2,3,1)/255     
        print("training images have been setup.")
        self.training_labels = y_train

        self.test_images = X_test
        test_len = len(self.test_images)
        self.test_images = self.test_images.reshape(test_len,3,32,32).transpose(0,2,3,1)/255     
        print("testing images have been setup.")     
        self.test_labels = y_test
        
    def next_batch(self,batch_size):
       
        X = self.training_images[self.i:self.i+batch_size].reshape(100,32,32,3)
        y = self.training_labels[self.i:self.i+batch_size]
        self.i = (self.i + batch_size) % len(self.training_images)
        return X, y
```

```python
import tensorflow as tf
```

```python
x = tf.placeholder(tf.float32,shape = [None,32,32,3])
y_true = tf.placeholder(tf.float32,shape = [None,26])
```

```python
hold_prob = tf.placeholder(tf.float32)
```

```python
def init_weights(shape):
    init_random_dist = tf.random_normal(shape,stddev=0.1)
    return tf.Variable(initial_value=init_random_dist)
```

```python
def init_bias(shape):
    init_random_bias = tf.constant(0.1,shape=shape)
    return tf.Variable(initial_value=init_random_bias)
```

```python
def conv2d(x,W):
    return tf.nn.conv2d(x,W,strides=[1,1,1,1],padding = 'SAME')
```

```python
def max_pool_2by2(x):
    return tf.nn.max_pool(x,[1,2,2,1],strides = [1,2,2,1],padding = 'SAME')
```

```python
def convolutional_layer(x,shape):
    W = init_weights(shape)
    b = init_bias([shape[3]])
    return tf.nn.relu(conv2d(x,W) + b)
```

```python
def normal_full_layer(x,size):
    input_size = int(x.get_shape()[1])
    W = init_weights([input_size,size])
    b = init_bias([size])
    return tf.matmul(x,W)+b
```

```python
convo_1= convolutional_layer(x,shape = [4,4,3,32])
convo_1_pooling = max_pool_2by2(convo_1)
```

```python
convo_2 = convolutional_layer(convo_1_pooling,shape = [4,4,32,64])
convo_2_pooling = max_pool_2by2(convo_2)
```

```python
convo2_flattened = tf.reshape(convo_2_pooling,[-1,4096])
```

```python
full_layer = normal_full_layer(convo2_flattened,size = 1024)
last_layer = tf.nn.relu(full_layer)
```

```python
dropout_layer = tf.nn.dropout(last_layer, hold_prob)
```

```python
y_pred = normal_full_layer(dropout_layer,size = 26)
```

```python
cross_entropy = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y_true,logits=y_pred))
```

```python
optimizer = tf.train.AdamOptimizer(0.001)
train = optimizer.minimize(cross_entropy)
```
